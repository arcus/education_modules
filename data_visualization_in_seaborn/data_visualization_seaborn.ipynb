## Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
breast_cancer_data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00451/dataR2.csv")
```

```python
breast_cancer_data.head()
```

## Scatterplots

```python
# this ensures we're starting with the default theme settings 
sns.set_theme()

# basic scatterplot
sns.relplot(data = breast_cancer_data, 
            x="Age", y="Glucose")

plt.savefig('media/seaborn_scatter_1.png')
```

```python
# using color for continuous variables
sns.relplot(data = breast_cancer_data, 
            x="Age", y="Glucose", hue="BMI")


plt.savefig('media/seaborn_scatter_2.png')
```

```python
# using color to show groups

# note that Classification is treated as a numeric variable in the data, but it should really be categorical
# convert the Classification column to categorical
breast_cancer_data["Classification"] = breast_cancer_data["Classification"].astype("category")


sns.relplot(data = breast_cancer_data, 
            x="Age", y="Glucose", 
            hue="Classification")

plt.savefig('media/seaborn_scatter_3.png')
```

```python
# distinguishing groups more clearly with color and shape
sns.relplot(data = breast_cancer_data, 
            x="Age", y="Glucose", 
            hue="Classification", style = "Classification")

plt.savefig('media/seaborn_scatter_4.png')
```

```python
# custom colors
sns.relplot(data = breast_cancer_data, 
            x="Age", y="Glucose", 
            hue="Classification", style = "Classification",
           palette = "colorblind")

plt.savefig('media/seaborn_scatter_5.png')

# to set the colorblind palette as default for all of our plots
sns.set_theme(palette="colorblind")
```

```python
# controlling figure aesthetics

# there are 5 preset seaborn themes: darkgrid, whitegrid, dark, white, and ticks
sns.set_style("white")


sns.relplot(data = breast_cancer_data, 
            x="Age", y="Glucose", 
            hue="Classification", style = "Classification")

plt.savefig('media/seaborn_scatter_6.png')
```

```python
# there are 4 different contexts available: notebook (default), paper, talk, and poster
sns.set_context("poster")

sns.relplot(data = breast_cancer_data, 
            x="Age", y="Glucose", 
            hue="Classification", style = "Classification")

plt.savefig('media/seaborn_scatter_7.png')

# set context back to notebook (the default)
sns.set_context("notebook")
```
